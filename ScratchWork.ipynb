```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import src.metadata_handler as mh

%matplotlib inline
```

```python
def get_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.cv.CV_BGR2RGB)
    return img

def show_img(img, title = ''):
    plt.subplot(111)
    plt.imshow(img)
    plt.suptitle(title)
    plt.show()
```

```python
dup_1 = get_image('data/second_sample/Hope Mountain/Hope Mountain-2015-1/01.01.15-02.01.15/11300135.JPG')
```
